```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly as py

# min year in your dataset
year = 1998
df=pd.read_csv('data/stmap.csv')

# your color-scale
scl = [[0.0, '#ffffff'],[0.2, '#b4a8ce'],[0.4, '#8573a9'],
       [0.6, '#7159a3'],[0.8, '#5732a1'],[1.0, '#2c0579']] # purples

data_slider = []
```

```python
for year in df['WEEK'].unique():
    df_segmented =  df[(df['WEEK']== year)]
```

```python
print(df_segmented)
```

```python
for col in df_segmented.columns:
    df_segmented[col] = df_segmented[col].astype(str)

    data_each_yr = dict(
                        type='choropleth',
                        locations = df_segmented['STATENAME'],
                        z=df_segmented['ACTIVITY LEVEL'],
                        locationmode='USA-states',
                        colorscale = scl,
#                         colorbar= {'title':'# Sightings'}
    )

    data_slider.append(data_each_yr)
```

```python
print(data_slider)
```

```python
steps = []
for i in range(len(data_slider)):
    step = dict(method='restyle',
                args=['visible', [False] * len(data_slider)],
                label='Week {}'.format(i ))
    step['args'][1][i] = True
    steps.append(step)

sliders = [dict(active=0, pad={"t": 1}, steps=steps)]

layout = dict(title ='CDC CLONED MAP', geo=dict(scope='usa',
                       projection={'type': 'albers usa'}),
              sliders=sliders)

fig = dict(data=data_slider, layout=layout)
py.offline.iplot(fig)
```
